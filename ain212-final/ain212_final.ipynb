```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests

# Load the dataset
file_id = '1DVgW8ryYhqIgW6IHLy4DCiVAE5ahmucr'
url = f'https://drive.google.com/uc?export=download&id={file_id}'
output_file = 'waterQuality1.csv'
response = requests.get(url)
if response.status_code == 200:
    # Write data into the file
    with open(output_file, 'wb') as f:
        f.write(response.content)
    df = pd.read_csv(output_file)

# First look at the data
print("Dataset shape:", df.shape)
print("\n" + "="*50)
print("First 5 rows:")
display(df.head())

print("\n" + "="*50)
print("Data types and non-null counts:")
print(df.info())

print("\n" + "="*50)
print("Basic statistics:")
display(df.describe())

print("\n" + "="*50)
print("Missing values per column:")
print(df.isnull().sum())

print("\n" + "="*50)
print("Target variable distribution (is_safe):")
print(df['is_safe'].value_counts())
```

```python
# Check unique non-numeric values in ammonia column
# First, let's try to convert and see which values fail

def check_non_numeric(column):
    non_numeric = []
    for val in column:
        try:
            float(val)
        except:
            non_numeric.append(val)
    return set(non_numeric)

print("Non-numeric values in 'ammonia':")
print(check_non_numeric(df['ammonia']))

print("\n" + "="*50)
print("Non-numeric values in 'is_safe':")
print(check_non_numeric(df['is_safe'].astype(str)))
```

```python

# STEP 2: DATA PREPROCESSING AND CLEANING

# 2.1 Check initial shape
print("Initial dataset shape:", df.shape)

# 2.2 Handle #NUM! errors
# Replace #NUM! with NaN first
df.replace('#NUM!', np.nan, inplace=True)

# Check how many rows contain NaN values
print("\nRows with missing values:", df.isnull().any(axis=1).sum())

# Drop rows with NaN values
df.dropna(inplace=True)

print("Shape after removing #NUM! rows:", df.shape)

# 2.3 Convert data types
# Convert ammonia to float
df['ammonia'] = df['ammonia'].astype(float)

# Convert is_safe to integer
df['is_safe'] = df['is_safe'].astype(int)

# 2.4 Verify the changes
print("\n" + "="*50)
print("Data types after conversion:")
print(df.dtypes)

print("\n" + "="*50)
print("Target variable distribution after cleaning:")
print(df['is_safe'].value_counts())

print("\n" + "="*50)
print("Basic statistics after cleaning:")
display(df.describe())
```

```python

# 2.5 Check for invalid negative values

print("Checking for negative values in each column:\n")

for col in df.columns:
    if col != 'is_safe':
        neg_count = (df[col] < 0).sum()
        if neg_count > 0:
            print(f"{col}: {neg_count} negative values")
            print(f"   Min value: {df[col].min()}")

# 2.6 Handle negative values

# Replace negative values with 0 (concentration cannot be negative)
for col in df.columns:
    if col != 'is_safe':
        df[col] = df[col].clip(lower=0)

print("\n" + "="*50)
print("After fixing negative values:")
print(f"Minimum value in ammonia: {df['ammonia'].min()}")

# 2.7 Check for outliers using IQR method

print("\n" + "="*50)
print("Outlier Analysis (values beyond 1.5*IQR):\n")

outlier_info = []

for col in df.columns:
    if col != 'is_safe':
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        outliers = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
        outlier_pct = (outliers / len(df)) * 100
        
        outlier_info.append({
            'Column': col,
            'Outliers': outliers,
            'Percentage': round(outlier_pct, 2)
        })

outlier_df = pd.DataFrame(outlier_info)
print(outlier_df.to_string(index=False))
```

```python
# 2.8 Visualize outliers with boxplots
fig, axes = plt.subplots(4, 5, figsize=(20, 16))
axes = axes.flatten()

for i, col in enumerate(df.columns[:-1]):
    axes[i].boxplot(df[col])
    axes[i].set_title(col)
    axes[i].set_ylabel('Value')

plt.suptitle('Boxplots of All Features (Outlier Visualization)', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()

# 2.9 Check class imbalance
print("\n" + "="*50)
print("CLASS IMBALANCE ANALYSIS")
print("="*50)

class_counts = df['is_safe'].value_counts()
class_percentages = df['is_safe'].value_counts(normalize=True) * 100

print(f"\nClass 0 (Not Safe): {class_counts[0]} samples ({class_percentages[0]:.2f}%)")
print(f"Class 1 (Safe):     {class_counts[1]} samples ({class_percentages[1]:.2f}%)")
print(f"\nImbalance Ratio: {class_counts[0] / class_counts[1]:.2f}:1")

# Visualize class distribution
fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(['Not Safe (0)', 'Safe (1)'], class_counts.values)

ax.set_ylabel('Number of Samples')
ax.set_title('Target Variable Distribution (Class Imbalance)')

for bar, count, pct in zip(bars, class_counts.values, class_percentages.values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 50, 
            f'{count}\n({pct:.1f}%)', ha='center', fontsize=11)

plt.tight_layout()
plt.show()

# 2.10 Feature Scaling
from sklearn.preprocessing import StandardScaler

# Separate features and target
X = df.drop('is_safe', axis=1)
y = df['is_safe']

# Store feature names
feature_names = X.columns.tolist()

# Apply StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame
X_scaled_df = pd.DataFrame(X_scaled, columns=feature_names)

print("\n" + "="*50)
print("FEATURE SCALING (Standardization)")
print("="*50)
print("\nBefore scaling (first 3 rows):")
display(X.head(3))

print("\nAfter scaling (first 3 rows):")
display(X_scaled_df.head(3))

print("\nScaled data statistics (should have mean≈0, std≈1):")
print(f"Mean of scaled features: {X_scaled.mean(axis=0).round(2)}")
print(f"Std of scaled features:  {X_scaled.std(axis=0).round(2)}")


# 2.11 Save preprocessed data
# Create preprocessed DataFrame
df_preprocessed = X_scaled_df.copy()
df_preprocessed['is_safe'] = y.values

# Save to CSV
df_preprocessed.to_csv('water_quality_preprocessed.csv', index=False)
print("\n" + "="*50)
print("Preprocessed data saved to 'water_quality_preprocessed.csv'")
print(f"Final shape: {df_preprocessed.shape}")
```

```python
# 3.5 CLUSTERING ANALYSIS
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage

print("\n" + "="*60)
print("CLUSTERING ANALYSIS")
print("="*60)

# Use scaled data for clustering
X_for_clustering = X_scaled

# 3.5.1 Finding Optimal K using Elbow Method and Silhouette Score

k_range = range(2, 11)
inertias = []
silhouette_scores = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_for_clustering)
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_for_clustering, kmeans.labels_))

# Plot Elbow Method and Silhouette Score
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Elbow Method
axes[0].plot(k_range, inertias, 'bo-', linewidth=2, markersize=8)
axes[0].set_xlabel('Number of Clusters (K)')
axes[0].set_ylabel('Inertia (Within-cluster Sum of Squares)')
axes[0].set_title('Elbow Method for Optimal K')
axes[0].grid(True, alpha=0.3)

# Silhouette Score
axes[1].plot(k_range, silhouette_scores, 'ro-', linewidth=2, markersize=8)
axes[1].set_xlabel('Number of Clusters (K)')
axes[1].set_ylabel('Silhouette Score')
axes[1].set_title('Silhouette Score for Different K Values')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nSilhouette Scores for each K:")
for k, score in zip(k_range, silhouette_scores):
    print(f"  K={k}: {score:.4f}")

# 3.5.2 K-Means Clustering with Optimal K
optimal_k = 3  # Based on elbow method and silhouette score

print(f"\n" + "="*50)
print(f"K-MEANS CLUSTERING (K={optimal_k})")
print("="*50)

kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
kmeans_labels = kmeans_final.fit_predict(X_for_clustering)

# Add cluster labels to dataframe
df['kmeans_cluster'] = kmeans_labels

print(f"\nCluster Distribution:")
print(df['kmeans_cluster'].value_counts().sort_index())

# Cluster characteristics
print(f"\nCluster Characteristics (Mean values):")
cluster_means = df.groupby('kmeans_cluster').mean()
display(cluster_means)

# 3.5.3 Visualize Clusters using PCA

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_for_clustering)

print(f"\nPCA Explained Variance Ratio: {pca.explained_variance_ratio_}")
print(f"Total Variance Explained: {sum(pca.explained_variance_ratio_)*100:.2f}%")

# Plot clusters in 2D PCA space
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# K-Means clusters
scatter1 = axes[0].scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans_labels, alpha=0.6, s=20)
axes[0].set_xlabel('First Principal Component')
axes[0].set_ylabel('Second Principal Component')
axes[0].set_title(f'K-Means Clustering (K={optimal_k})')
plt.colorbar(scatter1, ax=axes[0], label='Cluster')

# Actual labels (is_safe)
scatter2 = axes[1].scatter(X_pca[:, 0], X_pca[:, 1], c=y, alpha=0.6, s=20)
axes[1].set_xlabel('First Principal Component')
axes[1].set_ylabel('Second Principal Component')
axes[1].set_title('Actual Labels (is_safe)')
plt.colorbar(scatter2, ax=axes[1], label='is_safe')

plt.tight_layout()
plt.show()

# 3.5.4 Agglomerative Clustering

print(f"\n" + "="*50)
print("AGGLOMERATIVE CLUSTERING")
print("="*50)

# Dendrogram (using a sample for visualization)
sample_size = 500
np.random.seed(42)
sample_indices = np.random.choice(len(X_for_clustering), sample_size, replace=False)
X_sample = X_for_clustering[sample_indices]

plt.figure(figsize=(16, 6))
linkage_matrix = linkage(X_sample, method='ward')
dendrogram(linkage_matrix, truncate_mode='level', p=5)
plt.title('Hierarchical Clustering Dendrogram (Ward Linkage)')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.tight_layout()
plt.show()

# Agglomerative clustering with 3 clusters
agg_clustering = AgglomerativeClustering(n_clusters=3, linkage='ward')
agg_labels = agg_clustering.fit_predict(X_for_clustering)

df['agg_cluster'] = agg_labels

print(f"\nAgglomerative Cluster Distribution:")
print(df['agg_cluster'].value_counts().sort_index())

# 3.5.5 Compare Clustering with Actual Labels

print(f"\n" + "="*50)
print("CLUSTERING VS ACTUAL LABELS COMPARISON")
print("="*50)

# Crosstab of K-Means clusters vs is_safe
print("\nK-Means Clusters vs Actual Labels:")
kmeans_crosstab = pd.crosstab(df['kmeans_cluster'], df['is_safe'], margins=True)
kmeans_crosstab.columns = ['Not Safe', 'Safe', 'Total']
kmeans_crosstab.index = [f'Cluster {i}' if i != 'All' else 'Total' for i in kmeans_crosstab.index]
display(kmeans_crosstab)

# Percentage of safe water in each cluster
print("\nPercentage of Safe Water in Each K-Means Cluster:")
for cluster in range(optimal_k):
    cluster_data = df[df['kmeans_cluster'] == cluster]
    safe_pct = (cluster_data['is_safe'].sum() / len(cluster_data)) * 100
    print(f"  Cluster {cluster}: {safe_pct:.2f}% safe")
```

```python
# STEP 4: PREDICTIVE MODELLING
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

print("="*60)
print("STEP 4: PREDICTIVE MODELLING (Basic Approach)")
print("="*60)

# 4.1 Prepare Data

# Features and target
X = df.drop('is_safe', axis=1)
y = df['is_safe']

# 1. Split: Train+Val (%85) and Test (%15)
# Stratify removed explicitly
X_train_val, X_test, y_train_val, y_test = train_test_split(
    X, y, test_size=0.15, random_state=42
)

# 2. Split: Train (%70 of total) and Validation (%15 of total)
# 0.176 of 0.85 is roughly 0.15 of total.
X_train, X_val, y_train, y_val = train_test_split(
    X_train_val, y_train_val, test_size=0.176, random_state=42
)

# Scaling (Standardization)
# We fit on Training data only, then transform Val and Test
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

print("\n4.1 DATA SPLIT SUMMARY")
print("-"*40)
print(f"Training set:   {X_train.shape[0]} samples")
print(f"Validation set: {X_val.shape[0]} samples")
print(f"Test set:       {X_test.shape[0]} samples")

# 4.2 Model Training and Comparison

print("\n4.2 MODEL TRAINING")
print("-"*40)

# Define models
models = {
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42, n_estimators=100)
}

results = []

for name, model in models.items():
    print(f"Training {name}...")

    # Train on standard scaled data (No SMOTE)
    model.fit(X_train_scaled, y_train)

    # Predict on validation set
    y_val_pred = model.predict(X_val_scaled)

    # Calculate metrics
    accuracy = accuracy_score(y_val, y_val_pred)
    precision = precision_score(y_val, y_val_pred)
    recall = recall_score(y_val, y_val_pred)
    f1 = f1_score(y_val, y_val_pred)

    results.append({
        'Model': name,
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1
    })

# Results comparison table
print("\n" + "="*60)
print("MODEL COMPARISON (Validation Set)")
print("="*60)
results_df = pd.DataFrame(results)
# Sort by Accuracy since dataset is imbalanced but we are using basic metrics
results_df = results_df.sort_values('Accuracy', ascending=False)
display(results_df)

# 4.3 Hyperparameter Tuning (Random Forest)
# ============================================

print("\n4.3 HYPERPARAMETER TUNING (Random Forest)")
print("-"*40)

param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5]
}

rf = RandomForestClassifier(random_state=42)

grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=3, # Reduced CV folds for speed
    scoring='accuracy', # Changed to accuracy for basic approach
    n_jobs=-1,
    verbose=1
)

grid_search.fit(X_train_scaled, y_train)

print(f"\nBest Parameters: {grid_search.best_params_}")
print(f"Best Cross-Validation Accuracy: {grid_search.best_score_:.4f}")

# 4.4 Final Model Evaluation on Test Set

print("\n4.4 FINAL MODEL EVALUATION (Test Set)")
print("-"*40)

best_model = grid_search.best_estimator_

# Predict on test set
y_test_pred = best_model.predict(X_test_scaled)

# Calculate final metrics
final_accuracy = accuracy_score(y_test, y_test_pred)
final_precision = precision_score(y_test, y_test_pred)
final_recall = recall_score(y_test, y_test_pred)
final_f1 = f1_score(y_test, y_test_pred)

print(f"Test Set Results (Tuned Random Forest):")
print(f"  Accuracy:  {final_accuracy:.4f}")
print(f"  Precision: {final_precision:.4f}")
print(f"  Recall:    {final_recall:.4f}")
print(f"  F1-Score:  {final_f1:.4f}")
```
