# Questions (50p)

1. Read the csv file and create database. (10p)
2. Find the top 3 restaurants with the highest total revenue. (revenue = quantity x price) (5p)
3. Calculate the total quantity of dishes sold by category for each restaurant. (5p)
4. List the top 5 customer with highest total spending (also show total spending value) (5p)
5. Find the average rating for each restaurant. (5p)
6. Check for duplicate orders (same customer_id+restaurant_id+order_date). (5p)
7. Find the percentage of orders by delivery status. (7p)
8. Analyze whether specific cities contribute a significantly larger portion to the total revenue compared to others. (8p)

```python
full_df = pd.read_csv("Data/Foodpanda Analysis Dataset.csv")
conn = sqlite3.connect("foodpanda.db")
restaurant_columns = ['restaurant_name']
restaurants_df = full_df[restaurant_columns].drop_duplicates().reset_index(drop=True)
restaurants_df['restaurant_id'] = restaurants_df.index
restaurants_df.to_sql("Restaurant", conn, if_exists='replace', index=False)
customer_columns = [
    'customer_id', 'gender', 'age', 'city', 'signup_date', 
    'order_frequency', 'last_order_date', 'loyalty_points', 'churned'
]
customers_df = full_df[customer_columns].drop_duplicates(subset=['customer_id']).reset_index(drop=True)
customers_df.rename(columns={'age': 'age_group'}, inplace=True)
customers_df.to_sql("Customer", conn, if_exists='replace', index=False)
full_df = pd.merge(
    full_df, 
    restaurants_df,
    on='restaurant_name',
    how='left'
)
order_columns = [
    'order_id', 'customer_id', 'restaurant_id',
    'order_date', 'dish_name', 'category', 'quantity', 'price', 
    'payment_method', 'rating', 'rating_date', 'delivery_status'
]
orders_df = full_df[order_columns]
orders_df.to_sql("Orders", conn, if_exists='replace', index=False)
```

```python
query2 = """
SELECT
    r.restaurant_name,
    SUM(o.quantity * o.price) AS total_revenue
FROM
    Orders AS o
JOIN
    Restaurant AS r ON o.restaurant_id = r.restaurant_id
GROUP BY
    r.restaurant_name
ORDER BY
    total_revenue DESC
LIMIT
    3
"""

try:
    top_3_restaurants = pd.read_sql_query(query2, conn)

    print("Top 3 restaurants with the highest revenue:")
    display(top_3_restaurants)

except Exception as e:
    print(f"Query error: {e}")
```

```python
query3 = """
SELECT
    r.restaurant_name,
    o.category,
    SUM(o.quantity) AS total_quantity
FROM
    Orders AS o
JOIN
    Restaurant AS r ON o.restaurant_id = r.restaurant_id
GROUP BY
    r.restaurant_name, o.category
ORDER BY
    r.restaurant_name, total_quantity DESC
"""

try:
    category_sales = pd.read_sql_query(query3, conn)

    print("Total quantity of dishes sold by category for each restaurant:")
    display(category_sales)

except Exception as e:
    print(f"Query error: {e}")
```

```python
query4 = """
SELECT
    customer_id,
    SUM(quantity * price) AS total_spending
FROM
    Orders
GROUP BY
    customer_id
ORDER BY
    total_spending DESC
LIMIT
    5
"""

try:
    top_5_customers = pd.read_sql_query(query4, conn)

    print("Top 5 customer with highest total spending:")
    display(top_5_customers)

except Exception as e:
    print(f"Query error: {e}")
```

```python
query5 = """
SELECT
    r.restaurant_name,
    AVG(o.rating) AS average_rating
FROM
    Orders as o
JOIN
    Restaurant AS r ON o.restaurant_id = r.restaurant_id
WHERE
    o.rating IS NOT NULL
GROUP BY
    r.restaurant_name
ORDER BY
    average_rating DESC
"""

try:
    avg_ratings = pd.read_sql_query(query5, conn)

    print("Average rating for each restaurant:")
    display(avg_ratings)

except Exception as e:
    print(f"Query error: {e}")
```

```python
query6 = """
SELECT
    customer_id,
    restaurant_id,
    order_date,
    COUNT(*) AS duplicate_count
FROM
    Orders
GROUP BY
    customer_id, restaurant_id, order_date
HAVING
    duplicate_count > 1
"""

try:
    duplicate_orders = pd.read_sql_query(query6, conn)
    if len(duplicate_orders) > 0:
        print("Duplicate orders:")
        display(duplicate_orders)
    else:
        print("No duplicate orders found.")

except Exception as e:
    print(f"Query error: {e}")
```

```python
query7 = """
SELECT
    delivery_status,
    COUNT(*) AS status_count,
    (COUNT(*) * 100.0 / (SELECT COUNT(*) FROM Orders)) AS percentage
FROM
    Orders
GROUP BY
    delivery_status
ORDER BY
    percentage DESC
"""

try: 
    status_percentage = pd.read_sql_query(query7, conn)

    print("Percentage of orders by delivery status:")
    display(status_percentage)

except Exception as e:
    print(f"Query error: {e}")
```

```python
query8 = """
SELECT
    c.city,
    SUM(o.quantity * o.price) AS total_revenue,
    (SUM(o.quantity * o.price) * 100.0 / (SELECT SUM(quantity * price) FROM Orders)) AS revenue_percentage
FROM
    Orders AS o
JOIN
    Customer AS c ON o.customer_id = c.customer_id
GROUP BY
    c.city
ORDER BY
    total_revenue DESC
"""

try:
    city_revenue_analysis = pd.read_sql_query(query8, conn)

    print("Revenue Distribution and Percentage According to Cities")
    display(city_revenue_analysis)
except Exception as e:
    print(f"Query error: {e}")

finally:
    if 'conn' in locals() or 'conn' in globals():
        conn.close()
```
